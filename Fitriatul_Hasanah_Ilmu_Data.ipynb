**Import Data**

```python
import pandas as pd
# Membuat data penjualan online
data = {
 "Order_ID": [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008, 1009,
1010],
 "Tanggal": ["2025-01-01", "2025-01-02", "2025-01-03", "2025-01-04",
"2025-01-05",
 "2025-01-06", "2025-01-07", "2025-01-08", "2025-01-09",
"2025-01-10"],
 "Produk": ["Laptop A", "Mouse B", "Buku C", "Smartphone D", "Kaos E",
 "Celana F", "Keyboard G", "Novel H", "Headphone I", "Jaket J"],
 "Kategori": ["Elektronik", "Elektronik", "Buku", "Elektronik",
"Fashion",
 "Fashion", "Elektronik", "Buku", "Elektronik",
"Fashion"],
 "Harga": [12000000, 250000, 75000, 8000000, 150000,
 250000, 500000, 90000, 1000000, 350000],
 "Quantity": [1, 2, 3, 1, 2, 1, 1, 2, 1, 1]
}
```

```python
df = pd.DataFrame(data)
df["Pendapatan"] = df["Harga"] * df["Quantity"]
```

```python
file_path = "./penjualan_online.csv"
df.to_csv(file_path, index=False)
file_path
```

**Visualisasi Data**


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisasi Histogram untuk Harga Produk
plt.figure(figsize=(10, 6))
sns.histplot(df['Harga'], bins=10, kde=True, color='skyblue')
plt.title('Distribusi Harga Produk')
plt.xlabel('Harga')
plt.ylabel('Frekuensi')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisasi Boxplot untuk melihat outlier harga
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Harga'], color='lightgreen')
plt.title('Boxplot Harga Produk (Deteksi Outlier)')
plt.ylabel('Harga')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Hitung jumlah penjualan per kategori
penjualan_per_kategori = df.groupby('Kategori')['Quantity'].sum().reset_index()

# Visualisasi Bar Chart jumlah penjualan per kategori
plt.figure(figsize=(10, 6))
sns.barplot(x='Kategori', y='Quantity', data=penjualan_per_kategori, hue='Kategori', legend=False, palette='viridis')
plt.title('Jumlah Penjualan per Kategori')
plt.xlabel('Kategori Produk')
plt.ylabel('Jumlah Penjualan')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

**scatter plot antara Harga dan Quantity
untuk melihat korelasinya**


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisasi Scatter Plot antara Harga dan Quantity
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Harga', y='Quantity', data=df, hue='Kategori', size='Pendapatan', sizes=(20, 400), alpha=0.7)
plt.title('Scatter Plot Harga vs. Kuantitas (Korelasi)')
plt.xlabel('Harga Produk')
plt.ylabel('Kuantitas Terjual')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Kategori')
plt.tight_layout()
plt.show()
```

**Menganalisis distribusi harga produk**


* Distribusi Harga (Histogram): Sebagian besar produk memiliki harga yang relatif rendah. Ada beberapa produk dengan harga yang jauh lebih tinggi, menciptakan ekor panjang ke kanan pada distribusi.




* Deteksi Outlier (Boxplot): Boxplot mengkonfirmasi keberadaan beberapa outlier harga, yaitu produk-produk dengan harga yang sangat tinggi, yang jauh di atas rentang harga mayoritas produk lainnya. Produk-produk ini adalah 'Laptop A' dan 'Smartphone D'.

Analisis ini menunjukkan bahwa data harga produk memiliki sebaran yang tidak merata, dengan adanya produk-produk premium yang harganya sangat jauh berbeda dari produk lainnya.

**Menemukan kategori produk dengan pendapatan tertinggi**

```python
# 1. Menemukan kategori produk dengan pendapatan tertinggi
pendapatan_per_kategori = df.groupby('Kategori')['Pendapatan'].sum().reset_index()
kategori_tertinggi = pendapatan_per_kategori.loc[pendapatan_per_kategori['Pendapatan'].idxmax()]

print("Kategori produk dengan pendapatan tertinggi:")
display(kategori_tertinggi)
```

**Mengetahui pola penjualan berdasarkan tanggal**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pastikan kolom 'Tanggal' bertipe datetime
df['Tanggal'] = pd.to_datetime(df['Tanggal'])

# Menghitung total pendapatan per tanggal
penjualan_harian = df.groupby('Tanggal')['Pendapatan'].sum().reset_index()

# 2. Visualisasi pola penjualan berdasarkan tanggal
plt.figure(figsize=(12, 6))
sns.lineplot(x='Tanggal', y='Pendapatan', data=penjualan_harian, marker='o')
plt.title('Pola Penjualan Harian')
plt.xlabel('Tanggal')
plt.ylabel('Total Pendapatan')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
